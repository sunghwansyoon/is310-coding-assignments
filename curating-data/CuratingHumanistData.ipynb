```python
from bs4 import BeautifulSoup
import requests
import csv
import pandas as pd
```

```python
url = "https://humanist.kdl.kcl.ac.uk/Archives/"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
links = soup.find_all('a')
links = links[1:]
links
```

```python
converted_url = url + links[0].get('href')
sub_response = requests.get(converted_url)
sub_soup = BeautifulSoup(sub_response.text, "html.parser")
sub_links = sub_soup.find_all('a')
sub_links = sub_links[1:-1]
sub_links
```

```python
with open('Converted Text.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Link", "Text"])
    for sub_link in sub_links:
        if 'humanist' in sub_link.get('href'):
            volume_url = converted_url + sub_link.get('href')
            volume_response = requests.get(volume_url)
            volume_soup = BeautifulSoup(volume_response.text, "html.parser")
            writer.writerow([sub_link.get_text(), volume_soup.get_text()[:10000]])

convert = pd.read_csv('Converted Text.csv')
convert
```

```python
current_url = url + links[1].get('href')
sub_response = requests.get(current_url)
sub_soup = BeautifulSoup(sub_response.text, "html.parser")
sub_links = sub_soup.find_all('a')
sub_links = sub_links[1:]
sub_links
```

```python
with open('Current.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Link", "Text"])
    for sub_link in sub_links:
        if 'Humanist' in sub_link.get('href'):
            volume_url = current_url + sub_link.get('href')
            volume_response = requests.get(volume_url)
            volume_soup = BeautifulSoup(volume_response.text, "html.parser")
            writer.writerow([sub_link.get_text(), volume_soup.get_text()[:10000]])

current = pd.read_csv('Current.csv')
current
```

```python
virginia_url = "https://dhhumanist.org/"
sub_response = requests.get(virginia_url)
sub_soup = BeautifulSoup(sub_response.text, "html.parser")
sub_links = sub_soup.find_all('a')
# sub_links = sub_links[1:]
sorted_sub_links = []
for sub_link in sub_links:
    if '/volume/' in sub_link.get('href', ''):
        sorted_sub_links.append(sub_link)
sorted_sub_links
```

```python
with open('Virginia.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Link", "Text"])
    for sorted_sub_link in sorted_sub_links:
        volume_url = virginia_url + sorted_sub_link.get('href')
        sorted_sub_response = requests.get(volume_url)
        sorted_sub_soup = BeautifulSoup(sorted_sub_response.text, "html.parser")
        sorted_sub_links = sorted_sub_soup.find_all('a')
        sorted_sub_vol_links = []
        for sorted_link in sorted_sub_links:
            if '/volume/' in sorted_link.get('href'):
                sorted_sub_vol_links.append(sorted_link)
        for core_link in sorted_sub_vol_links:
            volume_url = volume_url + core_link.get('href')
            volume_response = requests.get(volume_url)
            volume_soup = BeautifulSoup(volume_response.text, "html.parser")
            writer.writerow([core_link.get_text(), volume_soup.get_text()[:10000]])

virginia = pd.read_csv('Virginia.csv')
virginia
```
