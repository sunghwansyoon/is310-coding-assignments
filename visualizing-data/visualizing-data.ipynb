```python
import subprocess
import sys

def install(package):
    subprocess.check_call([sys.executable, "-m", "pip", "install", package])

install('Altair')
import pandas as pd
import altair as alt

film_scripts = pd.read_csv('cleaned_pudding_data.csv')
film_scripts
```

```python
character_list_df = pd.read_csv('character_list5.csv', encoding='ISO-8859-1')
character_mapping_df = pd.read_csv('character_mapping.csv', encoding='ISO-8859-1')
metaddata_df = pd.read_csv('meta_data7.csv', encoding='ISO-8859-1')
```

```python
film_scripts.columns
```

```python
import altair as alt
```

```python
dialogue_df = pd.merge(character_list_df, metaddata_df, on='script_id', how="inner")
dialogue_df
```

```python
extended_dialogue_df = pd.merge(character_list_df, film_scripts, on='script_id', how="inner")
extended_dialogue_df
```

```python
grouped_ex_dialogue_df = extended_dialogue_df.groupby(['year', 'gender'])['words'].sum().reset_index()
grouped_ex_dialogue_df['date'] = grouped_ex_dialogue_df['year'].astype(str) + '-01-01'
grouped_ex_dialogue_df['date'] = pd.to_datetime(grouped_ex_dialogue_df['date'])
alt.Chart(grouped_ex_dialogue_df).mark_line().encode(
    x='date:T',
    y='words',
    color='gender'
).properties(
    title='Dialogue Over Time Colored By Gender (film_scripts)'
)
```

<h2> How could we tell if the amount of dialogue was increasing over time in movies? How might this influence the assessment about the breakdown of gender dialogue?

We can tell the amount of dialogue increased over time in movies by the graph provided above. Starting in 1990s, there was a dramatic increase of words in movies. Male characters had more dialogues than female characters over time.

<h2> How could test if there was any relationship between the film’s gross value and the amount of dialogue in the film?

```python
merge_df = pd.merge(dialogue_df, film_scripts[['gross (inflation-adjusted)']], left_on='gross', right_on='gross (inflation-adjusted)', how='inner')
merge_df
```

```python
grouped_merge_df = merge_df.groupby(['gross (inflation-adjusted)'])['words'].sum().reset_index()
alt.Chart(grouped_merge_df).mark_line().encode(
    x='gross (inflation-adjusted):Q',
    y='words',
).properties(
    title='Dialogue Over Gross Value (film_scripts)'
)
```

Gross value is lower when there were more dialogues in the film.
