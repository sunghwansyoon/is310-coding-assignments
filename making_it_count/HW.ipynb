```python
import pandas as pd
import altair as alt
```

<h3>First, make sure you have the Humanist listserv dataset and that you have dates associated with each volume

```python
humanist_vols = pd.read_csv('web_scraped_humanist_listserv_volumes.csv')
humanist_vols
```

```python
early_humanist = humanist_vols[humanist_vols['inferred_start_year'] < 1999]
early_humanist
```

```python
late_humanist = humanist_vols[humanist_vols['inferred_start_year'] >= 1999]
late_humanist
```

<h3>Second, would highly recommend using TF-IDF to help identify what is distinctive of these two time periods.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#save our texts to a list
documents = early_humanist.volume_text.tolist()

#Create a vectorizer
vectorizer = TfidfVectorizer(max_df=.7, min_df=1)

# Fit the vectorizer to our documents
transformed_documents = vectorizer.fit_transform(documents)

# Now get the top features for each document
transformed_documents_as_array = transformed_documents.toarray()

dates = early_humanist.inferred_start_year.tolist()
tfidf_results = []
for counter, doc in enumerate(transformed_documents_as_array):
    # construct a dataframe
    tf_idf_tuples = list(zip(vectorizer.get_feature_names_out(), doc))
    one_doc_as_df = pd.DataFrame.from_records(tf_idf_tuples, columns=['term', 'score']).sort_values(by='score', ascending=False).reset_index(drop=True)
    one_doc_as_df['inferred_start_year'] = dates[counter]
    tfidf_results.append(one_doc_as_df)

tfidf_df_early = pd.concat(tfidf_results)
tfidf_df_early.head(10)
```

```python
print(tfidf_df_early.term.unique())
```

```python
#save our texts to a list
documents = late_humanist.volume_text.tolist()

#Create a vectorizer
vectorizer = TfidfVectorizer(max_df=.7, min_df=1)

# Fit the vectorizer to our documents
transformed_documents = vectorizer.fit_transform(documents)

# Now get the top features for each document
transformed_documents_as_array = transformed_documents.toarray()

dates = late_humanist.inferred_start_year.tolist()
tfidf_results = []
for counter, doc in enumerate(transformed_documents_as_array):
    # construct a dataframe
    tf_idf_tuples = list(zip(vectorizer.get_feature_names_out(), doc))
    one_doc_as_df = pd.DataFrame.from_records(tf_idf_tuples, columns=['term', 'score']).sort_values(by='score', ascending=False).reset_index(drop=True)
    one_doc_as_df['inferred_start_year'] = dates[counter]
    tfidf_results.append(one_doc_as_df)

tfidf_df_late = pd.concat(tfidf_results)
tfidf_df_late.head(10)
```

<h3> Finally, you’ll need to visualize your results

```python
sample_early = tfidf_df_early.head(100)
sample_late = tfidf_df_late.head(100)
```

```python
# Calculate average TF-IDF score for each term
early_avg = sample_early.groupby('term')['score'].mean().reset_index()
late_avg = sample_late.groupby('term')['score'].mean().reset_index()

# Combine the two dataframes
combined = pd.concat([early_avg.assign(dataset='early'), late_avg.assign(dataset='late')])

# Create the chart
chart = alt.Chart(combined).mark_bar().encode(
    x='term:N',
    y='score:Q',
    color='dataset:N'
).properties(
    title='Average TF-IDF Score by Term'
)

chart
```
